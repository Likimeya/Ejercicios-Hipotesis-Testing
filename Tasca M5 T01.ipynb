Un valor P < o = 0,05 es considera estadísticament significatiu. Denota una forta evidència contra la hipòtesi nul·la, ja que hi ha menys d'un 5% de probabilitat que la nul·la sigui correcta. Per tant, rebutgem la hipòtesi nul·la i acceptem la hipòtesi alternativa.

Però si el valor P és inferior al vostre llindar de significació, tot i que es pot rebutjar la hipòtesi nul·la, no vol dir que hi hagi un 95% de probabilitat que la hipòtesi alternativa sigui certa. 

Un valor P > 0,05 no és estadísticament significatiu. Denota una forta evidència que la hipòtesi nul·la és certa. Així, conservem la hipòtesi nul·la i rebutgem la hipòtesi alternativa. No podem acceptar la hipòtesi nul·la; només podem rebutjar-lo o no rebutjar-lo. 

```python
import pandas as pd
import warnings
```

```python
from sklearn import svm
from sklearn import linear_model
```

```python
df = pd.read_csv("Lligues europees UEFA.csv", sep = ";")
df.head()
```

```python
Scorer = df["Top Team Scorer"].str.split('-', expand=True)
Scorer.head()
```

```python
df1 = pd.concat([df, Scorer], axis=1)
df1.head()
```

```python
df1.columns
```

```python
df1.drop(['Top Team Scorer'], axis=1)
```

```python
df1.rename({1: 'Scorer'}, axis=1, inplace=True)
```

```python
df1.rename({0: 'Top Team'}, axis=1, inplace=True)
```

```python
df1.head()
```

```python
df1.describe()
```

```python
import scipy.stats as stats
```

```python
df1.columns
```

```python
print(df1['Scorer'].dtype) 
```

```python
df1['Scorer'] = pd.to_numeric(df1['Scorer'], downcast="float")
```

```python
# busquem el promig dels golejadors dels equips
df1['Scorer'].mean()
```

```python
stats.ttest_1samp(a=df1['Scorer'], popmean=12.5918)
```

*com que pvalue=0.99 és major que 0,05 (alpha=5% grau de significancia) però no s'allunya tant podem dir que els resultat de la mostra sobre els màxims golejadors dels equips no és significativa, i per tant no podem refusar la Ho*
